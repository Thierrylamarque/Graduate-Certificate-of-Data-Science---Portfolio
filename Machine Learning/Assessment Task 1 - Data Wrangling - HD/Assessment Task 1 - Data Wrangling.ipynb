# Question: 5

#### Continue from question 3. Is there any association between heart rate and severity of heart disease? Explain your results from given dataset.

```python
# assign a slice of the data frame to a new variable to calculate association
heartrate_vs_state = df[['thalach','state']]

# calculate r by using Kendall method as our variables are continuous and ordinal (reference) 
corr = heartrate_vs_state.corr(method = 'kendall')

print(f"There is a weak negative association between heart rate and severity of heart disease \
with am r value being {format(corr.iloc[0,1],'.3f')}. The correlation method used is Kendall Tau correlation \
coefficient which is a good fit for measuring association of two variables where one is continuous \
and the other ordinal (Khamis, 2008).\n")

print("Reference: \nKhamis, H 2008, 'Measures of Association', Journal of Diagnostic Medical Sonography, Vol. 24, no. \
3,pp. 155-162, DOI: 10.1177/8756479308317006")
```
